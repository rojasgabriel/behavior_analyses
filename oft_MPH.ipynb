```python
def read_dat_file(filepath):
    import numpy as np
    data = np.genfromtxt(filepath,
                    skip_header=2,
                    skip_footer=1,
                    names=None,
                    dtype=None,
                    delimiter='\t',
                    usecols=(1,2))
    x_position = data[:,0]
    y_position = data[:,1]

    return x_position, y_position
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 4, figsize=(16, 12))

# Load the data
base_path = '/Users/gabriel/Library/CloudStorage/GoogleDrive-grojasbowe@gmail.com/My Drive/UCLA/ChurchlandLab/experiments/open-field-test_05-09-24/{}/{}.dat'
animal_ids = ['GRB032', 'GRB033', 'RS001', 'RS002']
conditions = ['baseline', 'test30', 'test70']

data = {}
titles = []

for animal_id in animal_ids:
    for condition in conditions:
        try:
            file_path = base_path.format(animal_id, animal_id + '_' + condition)
            data_key = animal_id + '_' + condition
            data[data_key] = read_dat_file(file_path)
            titles.append(data_key)
        except:
            continue

# Plot the data
for i, title in enumerate(titles):
    row = i % 3
    col = i // 3
    ax[row, col].plot(data[title][0], data[title][1], linewidth=0.4, color='k', alpha=1)
    ax[row, col].title.set_text(title)

plt.tight_layout()
plt.show()
```
