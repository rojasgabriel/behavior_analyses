```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def info(p_x):
    return np.log2(1 / p_x)


def entropy(x1, x2):
    return x1 * np.log2(1 / x1) + x2 * np.log2(1 / x2)
```

```python
info(0.9)
```

```python
x = np.arange(0, 1, 0.1)
ent = []
for i in x:
    ent.append(info(i))

plt.figure()
plt.plot(x, ent)
```
